```python
x["Meaning"]=""
```

```python
for i in range(len(x)):
    time.sleep(5)
    driver.get(f"{link}{x.iloc[i]['Word']}")
    try:
        temp_def = driver.find_element(by=By.XPATH, value = '//*[@id="page-content"]/div[2]/div[1]/div[2]/div/div[3]/div/div/div/div[3]/div/div[2]/div/div[2]/div').text
        temp_exmp = driver.find_element(by=By.CSS_SELECTOR,value=".ddef_b").find_elements(by=By.CSS_SELECTOR, value = ".ddef_b .dexamp")
        temp_exmp2='\n'.join(i.text for i in temp_exmp)
        x.loc[i,"Meaning"]=f"{temp_def}\n{temp_exmp2}"
        print(f"{temp_def}\n{temp_exmp2}")
    except:
        print(f"error {i}")
```
