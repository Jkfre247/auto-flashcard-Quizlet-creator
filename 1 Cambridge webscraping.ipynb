```python
import pandas as pd
import numpy as np
#import csv
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
def scraper(wordlist):
    #opening a link from cambridg
    path = "chromedriver.exe"
    driver = webdriver.Chrome(path)
    link = 'https://dictionary.cambridge.org/dictionary/english/'
    driver.get(link)
    wait=WebDriverWait(driver,20)
    time.sleep(3)
    #if a button for cookies pops up accept it
    try:
        wait.until(EC.visibility_of_element_located((By.ID,"onetrust-accept-btn-handler"))).click()
    except:
        pass
    for i in range(len(wordlist)):
        time.sleep(5)
        #creation of a custom link for link + word
        driver.get(f"{link}{wordlist.iloc[i][0]}")
        try:
            #finding the right content block
            base=driver.find_element(by=By.CLASS_NAME, value="entry-body")
            def_blocks=base.find_elements(by=By.CSS_SELECTOR, value=".ddef_block")
            fullspect = ""
            for block in def_blocks:
                #finding the right definition block
                temp_def=block.find_element(by=By.CSS_SELECTOR, value=".ddef_d").text
                temp_examp2=""
                try:
                    #finding the right block of examples
                    temp_examp = block.find_elements(by=By.CSS_SELECTOR, value = ".ddef_b .dexamp")
                    temp_examp2 = "\n".join(i.text for i in temp_examp)
                except:
                    print(i)
                    #putting it all together
                fullspect+=f'({temp_def})\n[{temp_examp2}]\n'
            print(f"\n\n{wordlist.iloc[i][0]}\n{fullspect}")
            wordlist.loc[i,"Meaning"]=fullspect
        except:
            print(i)
    
    driver.quit()
    return wordlist
```

```python
data= []
#loading files from a text file one line - word
wordlist = "words.txt"
file = open(wordlist, "r")
for i in file.readlines():
    data.append(i.split("\n")[0])
df = pd.DataFrame(data,columns=["Word"])
file.close()
df_2 = scraper(df)
df_2.to_csv("gotowy.csv",index=False)
```
